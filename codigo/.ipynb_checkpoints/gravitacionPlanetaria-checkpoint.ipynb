```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

##### Orbita Tierra al rededor del Sol

```python
# Constantes
G = 6.67430e-11  # Constante gravitacional (m^3 kg^-1 s^-2)
M_sun = 1.989e30  # Masa del Sol (kg)
M_earth = 5.972e24  # Masa de la Tierra (kg)
AU = 1.496e11  # Unidad Astronómica (distancia Tierra-Sol en metros)
year = 3.154e7  # Un año en segundos

# Condiciones iniciales
r0 = np.array([AU, 0])  # Posición inicial de la Tierra (1 UA en el eje x)
v0 = np.array([0, 29783])  # Velocidad inicial de la Tierra (m/s)

# Tiempo de simulación
dt = 24 * 60 * 60  # Paso de tiempo (1 día en segundos)
t_max = 365 * 24 * 60 * 60  # Tiempo total de la simulación (365 días en segundos)

# Función de fuerza gravitacional
def gravitational_force(r):
    r_magnitude = np.linalg.norm(r)
    force_magnitude = G * M_sun * M_earth / r_magnitude**2
    force_direction = -r / r_magnitude
    return force_magnitude * force_direction

# Inicialización de arrays
t_values = np.arange(0, t_max, dt)
r_values = np.zeros((len(t_values), 2))
v_values = np.zeros((len(t_values), 2))

# Condiciones iniciales
r_values[0] = r0
v_values[0] = v0

# Método de Verlet para la integración numérica
for i in range(1, len(t_values)):
    r = r_values[i-1]
    v = v_values[i-1]
    
    # Calcula la fuerza y la aceleración
    F = gravitational_force(r)
    a = F / M_earth
    
    # Actualiza la posición usando el método de Verlet
    r_new = r + v * dt + 0.5 * a * dt**2
    
    # Calcula la nueva fuerza y aceleración en la nueva posición
    F_new = gravitational_force(r_new)
    a_new = F_new / M_earth
    
    # Actualiza la velocidad
    v_new = v + 0.5 * (a + a_new) * dt
    
    # Guarda los valores actualizados
    r_values[i] = r_new
    v_values[i] = v_new

# Configuración de la animación
fig, ax = plt.subplots(figsize=(8, 8))
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
ax.set_xlabel('Distancia en el eje x (UA)')
ax.set_ylabel('Distancia en el eje y (UA)')
ax.set_title('Simulación de la órbita de la Tierra alrededor del Sol')
ax.grid(True)
ax.set_aspect('equal', adjustable='box')
line, = ax.plot([], [], 'b-', label='Trayectoria de la Tierra')
point, = ax.plot([], [], 'bo', label='Tierra')
sun, = ax.plot(0, 0, 'yo', label='Sol')
ax.legend()

# Función de inicialización
def init():
    line.set_data([], [])
    point.set_data([], [])
    return line, point

# Función de actualización
def update(frame):
    xdata = r_values[:frame+1, 0] / AU
    ydata = r_values[:frame+1, 1] / AU
    line.set_data(xdata, ydata)
    point.set_data(xdata[-1], ydata[-1])
    return line, point

# Creación de la animación
frames_per_second = len(t_values) / 3.65
ani = FuncAnimation(fig, update, frames=len(t_values), init_func=init, blit=True, interval=1000/frames_per_second)

# Guardar la animación (opcional)
ani.save('orbita_tierra.mp4', writer='ffmpeg', fps=frames_per_second)
```

##### Orbita de MErcurio, Venus, Tierra, MArte

```python
# Constantes
G = 6.67430e-11  # Constante gravitacional (m^3 kg^-1 s^-2)
M_sun = 1.989e30  # Masa del Sol (kg)
AU = 1.496e11  # Unidad Astronómica (distancia Tierra-Sol en metros)
year = 3.154e7  # Un año en segundos

# Datos de los planetas: [masa (kg), distancia inicial (AU), velocidad inicial (m/s), periodo orbital (días)]
planetas = {
    'Mercurio': [3.301e23, 0.387, 47360, 88],
    'Venus': [4.867e24, 0.723, 35020, 225],
    'Tierra': [5.972e24, 1.000, 29783, 365],
    'Marte': [0.6417e24, 1.524, 24007, 687]
}

# Tiempo de simulación
dt = 24 * 60 * 60  # Paso de tiempo (1 día en segundos)
t_max = 365 * 24 * 60 * 60  # Tiempo total de la simulación (365 días en segundos)

# Función de fuerza gravitacional
def gravitational_force(r, M_planet):
    r_magnitude = np.linalg.norm(r)
    force_magnitude = G * M_sun * M_planet / r_magnitude**2
    force_direction = -r / r_magnitude
    return force_magnitude * force_direction

# Inicialización de arrays
t_values = np.arange(0, t_max, dt)
r_values = {planet: np.zeros((len(t_values), 2)) for planet in planetas}
v_values = {planet: np.zeros((len(t_values), 2)) for planet in planetas}

# Condiciones iniciales
for planet, (mass, dist, vel, _) in planetas.items():
    r_values[planet][0] = np.array([dist * AU, 0])
    v_values[planet][0] = np.array([0, vel])

# Método de Verlet para la integración numérica
for i in range(1, len(t_values)):
    for planet, (mass, _, _, _) in planetas.items():
        r = r_values[planet][i-1]
        v = v_values[planet][i-1]
        
        # Calcula la fuerza y la aceleración
        F = gravitational_force(r, mass)
        a = F / mass
        
        # Actualiza la posición usando el método de Verlet
        r_new = r + v * dt + 0.5 * a * dt**2
        
        # Calcula la nueva fuerza y aceleración en la nueva posición
        F_new = gravitational_force(r_new, mass)
        a_new = F_new / mass
        
        # Actualiza la velocidad
        v_new = v + 0.5 * (a + a_new) * dt
        
        # Guarda los valores actualizados
        r_values[planet][i] = r_new
        v_values[planet][i] = v_new

# Cálculo de vueltas completadas por cada planeta
vueltas = {planet: 365 / periodo for planet, (_, _, _, periodo) in planetas.items()}

# Configuración de la animación
fig, ax = plt.subplots(figsize=(8, 8))
ax.set_xlim(-1.6, 1.6)
ax.set_ylim(-1.6, 1.6)
ax.set_xlabel('Distancia en el eje x (UA)')
ax.set_ylabel('Distancia en el eje y (UA)')
ax.set_title('Simulación de las órbitas de los planetas alrededor del Sol')
ax.grid(True)
ax.set_aspect('equal', adjustable='box')

lines = {planet: ax.plot([], [], label=f'{planet} ({vueltas[planet]:.2f} vueltas)')[0] for planet in planetas}
points = {planet: ax.plot([], [], 'o')[0] for planet in planetas}
ax.plot(0, 0, 'yo', label='Sol')
ax.legend()

# Función de inicialización
def init():
    for line in lines.values():
        line.set_data([], [])
    for point in points.values():
        point.set_data([], [])
    return list(lines.values()) + list(points.values())

# Función de actualización
def update(frame):
    for planet in planetas:
        xdata = r_values[planet][:frame+1, 0] / AU
        ydata = r_values[planet][:frame+1, 1] / AU
        lines[planet].set_data(xdata, ydata)
        points[planet].set_data(xdata[-1], ydata[-1])
    return list(lines.values()) + list(points.values())

# Creación de la animación
frames_per_second = len(t_values) / 3.65
ani = FuncAnimation(fig, update, frames=len(t_values), init_func=init, blit=True, interval=1000/frames_per_second)

# Guardar la animación (opcional)
ani.save('orbita_planetas.mp4', writer='ffmpeg', fps=frames_per_second)
```
